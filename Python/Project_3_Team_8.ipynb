```python
#create a heatmap that will show the market value of each team
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#loop through api_response and create a dataframe with the team name and market value
market_cap = pd.DataFrame(columns=['Team', 'Market Value'])
for team in api_response['teams']:
    market_cap = market_cap.append({'Team': team['clubName'], 'Market Value': team['marketValueUnformatted']}, ignore_index=True)

```

```python
#find a way to locate where each team is from and add that to the dataframe
market_cap['Country'] = ''
for index, row in market_cap.iterrows():
    team_name = row['Team']
    querystring = {"query":team_name,"domain":"de"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    api_response = response.json()
    market_cap.loc[market_cap['Team'] == team_name, 'Country'] = api_response['teams'][0]['countryName']
```
